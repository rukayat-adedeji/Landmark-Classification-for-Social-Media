### Description of the Chosen Procedure for Preprocessing the Data
The preprocessing of the data involves several key steps, including resizing, cropping, normalization, and augmentation. These steps are implemented using the torchvision.transforms module, which provides a variety of transformation functions for image data.

**Image Resizing and Cropping**
The procedure for resizing and cropping the images is as follows:

1. *Resize*: All images are first resized to 256x256 pixels using T.Resize(256). This ensures that all images have a uniform size, which simplifies subsequent processing steps.
2. *Center/Random Crop*: After resizing, the images are cropped to 224x224 pixels using T.CenterCrop(224) for test and calidation data or T.RandomCrop(224). Center cropping the test/validation data is chosen to remove the outer edges of the images uniformly, ensuring the main content of the image is retained. A random crop of the train image is chosen to allow the model be trained on various field area of the same image from epoch to epoch.
The chosen size for the input tensor is 224x224 pixels. This size is standard for many convolutional neural networks (CNNs), such as those used in popular architectures like ResNet and VGG. The reason for choosing this size is that it provides a good balance between retaining sufficient detail in the images and reducing the computational load during training and inference.

**Data Augmentation**
Yes, the dataset is augmented during the training phase. Data augmentation helps to increase the diversity of the training data without actually collecting new data, which can improve the robustness and generalization of the trained model. The augmentation techniques applied include:

3. *Random Affine Transformations*: T.RandomAffine(translate=(0.1, 0.1), degrees=0, interpolation=T.InterpolationMode.BILINEAR) is used to apply random affine transformations with a translation up to 10% of the image size. This helps the model to be invariant to small translations in the image.
4. *Color Jitter*: T.ColorJitter(brightness=(0.5, 1), contrast=(0.5, 1)) is used to randomly change the brightness and contrast of the images. This helps the model to be robust to different lighting conditions.
5. *Random Horizontal Flip*: T.RandomHorizontalFlip(p=0.2) is used to randomly flip the images horizontally with a probability of 0.2. This helps the model to learn features that are invariant to horizontal flipping.
The validation and test datasets are not augmented. For these datasets, only resizing, center cropping, and normalization are applied to ensure that the evaluation is consistent and not influenced by random transformations.

**Normalization**
Finally, the images are normalized using T.Normalize(mean, std), where mean and std are the mean and standard deviation of the dataset, respectively. Normalization helps to standardize the input data, making the training process more stable and faster. The normalization parameters are computed based on the entire dataset to ensure consistency.

## Finding Optimum lr and dropout rate

````python
import torch
import numpy as np
from torch.optim.lr_scheduler import LambdaLR
from tqdm import tqdm
import torch.optim as optim
import copy
import matplotlib.pyplot as plt

def lr_finder(min_lr, max_lr, n_steps, loss, model, data_loaders):
    
    # Save initial weights so we can restore them at the end
    torch.save(model.state_dict(), "__weights_backup")
    
    # specify optimizer
    optimizer = optim.SGD(model.parameters(), lr=min_lr)

    # We create a learning rate scheduler that increases the learning
    # rate at every batch.
    # Find the factor where min_lr r**(n_steps-1) = max_lr
    r = np.power(max_lr / min_lr, 1 / (n_steps - 1))

    def new_lr(epoch):
        """
        This should return the *factor* by which the initial learning
        rate must be multipled for to get the desired learning rate
        """
        return r ** epoch

    # This scheduler increases the learning rate by a constanct factor (r)
    # at every iteration
    lr_scheduler = LambdaLR(optimizer, new_lr)

    # Set the model in training mode
    # (so all layers that behave differently between training and evaluation,
    # like batchnorm and dropout, will select their training behavior)
    model.train()
    if torch.cuda.is_available():
            model.cuda()
    # Loop over the training data
    losses = {}
    train_loss = 0.0

    for batch_idx, (data, target) in tqdm(
        enumerate(data_loaders["train"]),
        desc="Training",
        total=len(data_loaders["train"]),
        leave=True,
        ncols=80,
    ):
        # move data to GPU if available
        if torch.cuda.is_available():
            data, target = data.cuda(), target.cuda()

        # 1. clear the gradients of all optimized variables
        optimizer.zero_grad()  # -
        # 2. forward pass: compute predicted outputs by passing inputs to the model
        output = model(data)  # =
        # 3. calculate the loss
        loss_value = loss(output, target)  # =
        # 4. backward pass: compute gradient of the loss with respect to model parameters
        loss_value.backward()  # -
        # 5. perform a single optimization step (parameter update)
        optimizer.step()  # -

        train_loss = train_loss + (
            (1 / (batch_idx + 1)) * (loss_value.data.item() - train_loss)
        )

        losses[lr_scheduler.get_last_lr()[0]] = train_loss

        # Stop if the loss gets too big
        if train_loss / min(losses.values()) > 10:
            break

        if batch_idx == n_steps - 1:
            break
        else:
            # Increase the learning rate for the next iteration
            lr_scheduler.step()
    
    # Restore model to its initial state
    model.load_state_dict(torch.load('__weights_backup'))
    
    return losses
````

````python
import torch.nn as nn
batch_size = 32
valid_size = 0.2
num_workers = multiprocessing.cpu_count()

lr_dataloaders = get_data_loaders(batch_size, valid_size, num_workers=num_workers)
lr_model = MyModel(num_classes=50, dropout=0.2)
# Range  and number of steps for the learning rate
min_lr = 1e-5
max_lr = 1
n_steps = min(len(data_loaders['train']), 200)

# specify loss function (categorical cross-entropy)
lr_loss = nn.CrossEntropyLoss()

losses = lr_finder(min_lr, max_lr, n_steps, lr_loss, lr_model, lr_dataloaders)

# Plot the results
plt.plot(losses.keys(), losses.values())
plt.xscale("log")
plt.ylabel("loss")
plt.xlabel("learning rate (log scale)")

# Adjust the range on the y-axis to see things more clearly
plt.xlim([1e-4, None])
plt.ylim([min(losses.values()), np.percentile(list(losses.values()), 97)])
````
